# One Hot Encoding Or Dummy Variable

```python
import pandas as pd
from pandas import get_dummies
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('loan_risk_prediction.csv', encoding='latin1')
```

```python
data
```

```python
data.columns
```

```python
gender = get_dummies(data['Gender'])
data = pd.concat([data, gender], axis=1)
data = data.drop(['Gender'], axis=1)

education = get_dummies(data['Education'])
data = pd.concat([education,data],axis=1)
data = data.drop(['Education'], axis=1)
```

```python
encoder =LabelEncoder()
status = encoder.fit_transform(data[['Loan_Status']])
```

```python
data['Loan_Status'] = status
data.head()
```

```python
features = ['Graduate','Not Graduate','Dependents','ApplicantIncome','CoapplicantIncome']
```

```python
data['Dependents'] = data['Dependents'].fillna(data.Dependents.median())
```

```python
data.Dependents.isna().sum()
```

```python
labels = ['Loan_Status']
```

```python
data[labels]
```

```python
model = LinearRegression()
```

```python
model.fit(data[features],data[labels])
```

```python
model.predict([[1,0,0.0,5849,0.0]])
```
